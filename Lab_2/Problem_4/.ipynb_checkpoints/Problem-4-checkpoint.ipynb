# Problem 4: Convergence of Forces with respect to K-Points

Using PWscf, calculate the force on a Ge atom displaced +0.05 in the z direction (in fractional coordinates) as a function of k-point grid size. Keep all other parameters fixed. Record your relevant conditions (lattice parameter, cutoffs, etc).

```python
import numpy as np
import subprocess
import matplotlib.pyplot as plt 
```

```python
import matplotlib_inline.backend_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

### Loading in our helper files 

```python
def force(x, y, z):
    """ returns the magnitude of the force """
    array = np.array([float(x),float(y),float(z)])
    inner_prod = np.dot(array, array)
    return np.sqrt(inner_prod)


def parse_output(outfile):
    """ Parses the quantum espresso output file
        Returns a dictionary with relevants in key-value pairs
        such as {'energy': energy_value}
    """
    
    with open(outfile, 'r') as outf:
        for line in outf:
            if (line.lower().startswith('     lattice parameter (alat)')):
                lattice_constant = float(line.split()[-2]) * 0.529177
            if (line.lower().startswith('!    total energy')):
                total_energy = float(line.split()[-2]) * 13.605698066
            if (line.lower().startswith('     atom    2 type  1   force')):
                test_string = line.split()
                x = test_string[-1]
                y = test_string[-2]
                z = test_string[-3]
                atom_force = force(x,y,z)
    
    result = {'energy': total_energy, 'lattice': lattice_constant, 'force': atom_force}
    return result 
```

## Update Function
We begin with the assumption that we have a working input file for our problem. Then at each iteration, all we want to change is the k-points value. This function performs this substitution by writing in the input file as text, parsing through until it gets to the relevant line, writes down a new string, and saves it. 

```python
def k_points_subs(file, k_points, offset = '0'):
    """ opens input file file and changes value for k-points """
    
    # check for proper input 
    assert type(k_points) == 'int' or 'numpy.int64', 'Provide k-points input as int or numpy.int64' 
    k_points = str(k_points)
    
    # generate k_points string
    k_string = '{} {} {} {} {} {}\n'.format(k_points, k_points,\
                                          k_points, offset, offset, offset) 
    
    # open the file 
    with open(file,'r') as input_file:
        lines = input_file.readlines()
    with open(file, 'w') as input_file:
        for line in lines:
            # checks if line corresponds to "K_POINTS AUTOMATIC" line 
            if line.split()[0] == 'K_POINTS':
                # if so: write same line again
                input_file.write(line)
                # write the k_points string 
                input_file.write(k_string)
            else:
                input_file.write(line)                
```

### Defining the k-points array values

```python
k_points_array = np.arange(1,15,1)
```

### Writing everything in a single function call

```python
def problem_4(k_points_array):
    """ performs problem 4 calculations 
        finds energy as a function of k-points 
    """
    
    # declare force array of same size as cut off array 
    force_array = np.zeros(len(k_points_array))
    
    for i in range(len(k_points_array)):
        # update input file 
        k_points_subs("Ge.scf.in", k_points_array[i])
        
        # call pw.x 
        subprocess.run('pw.x -in Ge.scf.in > ge.scf.out', shell=True)
        
        # parse output file 
        result = parse_output('ge.scf.out')
        
        # get force and append to array 
        force_array[i] = result['force']
        
    return force_array
```

```python
output = problem_4(k_points_array)
```

## Plotting

```python
plt.scatter(k_points_array, output)
plt.grid() 
plt.xlabel('K Points')
plt.ylabel('Force in Ry/Bohr')
plt.title('Ge: Force Convergence wrt K Points')
plt.savefig('force_vs_k_points.pdf')
```

## Checking for convergence
A condition on convergence is given by
$$
    10\frac{\text{meV}}{\text{Angstrom}} = 3.909 \times 10^{-4}\frac{\text{Ry}}{\text{Bohr}}
$$
where convergence to $X$ Rydbergs is given when $$E_\text{true} - E_n = X$$ For our case we can take $E_\text{true} \approx E_\text{final}$. 

```python
def force_diff(output_array):
    E_true = output_array[-1]
    force_diff_array = np.zeros(len(output_array))
    for i in range(len(output_array)):
        force_diff_array[i] = E_true - output_array[i]
    return force_diff_array
```

```python
convergence = force_diff(output)
convergence
```

### Extract convergence point 

```python
def convergence_point(k_points, convergence_array):
    """ find point in k_points array such that we get a 
        force within convergence threshold 
    """
    threshold = 3.909e-4 #Ry/Bohr
    counter = 0
    while (np.abs(convergence_array[counter]) > threshold):
        counter += 1
    return k_points[counter]
```

```python
convergence_point(k_points_array, convergence)
```
